```python
base_path = "/Volumes/setorfarmaceutico/bronze/bronze_data/" 
```

#### Criação da Tabela "Princípios Ativos"

```python
# 1) Princípios Ativos
df_principios = (spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{base_path}/principios_ativos.csv"))

df_principios.write.format("delta").mode("overwrite") \
    .saveAsTable("setorfarmaceutico.bronze.principios_ativos")
```

%md
#### Criação da Tabela "Produtos Fabricados"

```python
# 2) Produtos Fabricados
df_produtos = (spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{base_path}/produtos_fabricados.csv"))

df_produtos.write.format("delta").mode("overwrite") \
    .saveAsTable("setorfarmaceutico.bronze.produtos_fabricados")
```

%md
#### Criação da Tabela "Vendas Produtos Finais"

```python
# 3) Vendas Produtos Finais 
df_vendas = (spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{base_path}/vendas_produtos_finais.csv"))

df_vendas.write.format("delta").mode("overwrite") \
    .saveAsTable("setorfarmaceutico.bronze.vendas_produtos_finais")
```
